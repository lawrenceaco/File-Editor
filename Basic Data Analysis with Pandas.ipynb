**pandas** is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool,
built on top of the Python programming language.

```python
import pandas as pd
```

```python
dataset = pd.read_csv('Cities.csv')
```

**Filter rows**

```python
dataset[dataset.country == 'Ukraine']
```

```python
dataset[dataset['country'] == 'Ukraine']
```

```python
dataset[(dataset.country == 'Ukraine') & (dataset.temperature > 10)]
```

```python
dataset[dataset.temperature > 10]
```

```python
dataset[~(dataset.temperature > 10)]
```

```python
dataset[dataset.city.str.startswith('B')]
```

```python
dataset[dataset.city.str.startswith('Be')]
```

```python
dataset[dataset.country.str.endswith('a')]
```

```python
dataset[dataset.country.str.endswith('a')].drop_duplicates(subset='country')
```

```python
dataset[dataset.country.str.endswith('a')].drop_duplicates(subset='country')['country']
```

```python
dataset[dataset.city.str.contains('x')]
```

```python
dataset[(dataset.city.str.startswith('B')) & (dataset.temperature > 10)]
```

```python
dataset[(dataset.city.str.startswith('B')) | (dataset.temperature > 10)]
```

```python
dataset[dataset.country.isin(['Ukraine', 'United Kingdom','Romania'])]
```

**Filter column and row**

```python
dataset[['country','city','temperature']][(dataset.city.str.startswith('B')) & (dataset.temperature > 10)]
```

**Sort rows**

```python
dataset.sort_values(by='temperature')
```

```python
dataset.sort_values(by='temperature', ascending = False)
```

```python
dataset[['country','city','temperature']].sort_values(by=['country','temperature'])
```

```python
dataset.sort_values(by=['country','temperature'], ascending=[True,False])
```

```python
dataset[['country','city','temperature']][dataset.temperature > 10].sort_values(by='temperature')
```

**Use aggregate functions**

```python
sum(dataset.temperature)
```

```python
max(dataset.temperature)
```

```python
dataset[dataset.temperature == max(dataset.temperature)]
```

```python
dataset[dataset.temperature == min(dataset.temperature)]
```

```python
#Average temperature in Europe
sum(dataset.temperature) / len(dataset)
```

```python
from statistics import mean

mean(dataset.temperature)
```

**Summarizing Records using groupby**

```python
dataset.groupby(by='country')[['city']].count()
```

```python
dataset.groupby(by='country')[['city']].count().reset_index()
```

```python
dataset.groupby(by='country')[['temperature']].max()
```

```python
import numpy as np

dataset.groupby(by='country')[['temperature']].agg({min,max,np.mean})
```

```python
dataset.groupby(by='country').agg({'latitude':min,'longitude':max})
```
